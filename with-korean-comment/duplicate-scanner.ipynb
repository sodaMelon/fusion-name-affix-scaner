```python
#click 1nd play button (▶)
import re

from google.colab import files
Uploded = files.upload()

# 입력 데이터: ↓업로드 파일에 맞춰 이름바꿔주세요
with open('sample2.txt', 'r', encoding='utf-8') as file:
    data = file.read()


# press upload file button below.
```

```python
# fusionPrefix 및 fusionSuffix 추출을 위한 정규식 패턴
prefix_pattern = r'fusionPrefix:\s*"([^"]+)"'
suffix_pattern = r'fusionSuffix:\s*"([^"]+)"'

# 패턴을 사용하여 fusionPrefix 및 fusionSuffix 값을 모두 추출
fusion_prefixes = re.findall(prefix_pattern, data)
fusion_suffixes = re.findall(suffix_pattern, data)

# 중복된 fusionPrefix 값을 확인
duplicate_prefixes = set()
seen_prefixes = set()

for prefix in fusion_prefixes:
    if prefix in seen_prefixes:
        duplicate_prefixes.add(prefix)
    else:
        seen_prefixes.add(prefix)

# 중복된 fusionSuffix 값을 확인
duplicate_suffixes = set()
seen_suffixes = set()

for suffix in fusion_suffixes:
    if suffix in seen_suffixes:
        duplicate_suffixes.add(suffix)
    else:
        seen_suffixes.add(suffix)

# 결과 출력
if duplicate_prefixes:
    print("중복된 fusionPrefix 값이 있습니다:", duplicate_prefixes)
else:
    print("중복된 fusionPrefix 값이 없습니다.")

if duplicate_suffixes:
    print("중복된 fusionSuffix 값이 있습니다:", duplicate_suffixes)
else:
    print("중복된 fusionSuffix 값이 없습니다.")
```
